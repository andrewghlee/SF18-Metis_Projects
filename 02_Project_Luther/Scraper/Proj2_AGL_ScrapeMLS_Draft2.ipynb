```python
from __future__ import print_function, division
import requests
import pandas as pd
from bs4 import BeautifulSoup

url = 'https://www.mlslistings.com/Search/Result/d91d5243-c98f-4ce2-9f90-4b362b047ff1/1?view=list'
response = requests.get(url)
print("Check status code for MLS Listings - San Francisco (Sold): ",response.status_code)
url = 'https://www.mlslistings.com/Search/Result/a28f818b-9497-4094-99b5-8b577f79e62b/1?view=list'
response = requests.get(url)
print("Check status code for MLS Listings - San Francisco (Current Listing): ",response.status_code)
```

```python
# Create empty arrays for scraping
titles = []
prices = []
mlsnum = []
bednum = []
batnum = []
status = []
sqfoot = []
buildt = []
broker = []

# Function to scrape data from mlslistings.com - San Francisco (Sold)
def mls_sfsold_data(pagenum):
    
    # Setup a loop to go through the pages
    for i in range(1, pagenum+1):
        try:
            url = 'https://www.mlslistings.com/Search/Result/d91d5243-c98f-4ce2-9f90-4b362b047ff1/{0}?view=list'.format(i)
            response = requests.get(url)
            page = response.text
            soup = BeautifulSoup(page,"lxml")
        except:
            break
        
        # For loop to grab addresses
        for j in range(1,81,2):
            try:
                listing_title = soup.find_all(class_='search-nav-link')[j]['title']
                titles.append(listing_title)
            except:
                break
        
        # For loop to grab prices
        for k in range(0,40):
            try:
                listing_price = soup.find_all(class_='font-weight-bold listing-price d-block pull-left pr-25')[k].text
                try:
                    listing_price = "{0:.2f}".format(round(float(listing_price.replace('$','').replace(',','')),2))
                except:
                    listing_price = "{0:.2f}".format(float(0))
                prices.append(listing_price)
            except:
                break
        
        # For loop to grab MLS Numbers
        for k in range(0,40):
            try:
                mls_label = soup.find_all(class_='info-item-label d-block pull-left font-weight-bold')[k].text
                mlsnum.append(mls_label)
            except:
                break
        
        # For loop to grab the number of bedrooms
        for k in range(0,40):
            try:
                num_beds = soup.find_all(class_='listing-info-item font-size-sm line-height-base d-block pull-left pr-50 listing-beds')[k].find_all(class_='font-weight-bold info-item-value d-block pull-left pr-25')[0].text
                try:
                    num_beds = int(num_beds)
                except:
                    num_beds = int(0)                    
                bednum.append(num_beds)
            except:
                break
        
        # For loop to grab the number of bathrooms
        for k in range(0,40):
            try:
                num_baths = soup.find_all(class_='listing-info-item font-size-sm line-height-base d-block pull-left pr-50 listing-baths')[k].find_all(class_='font-weight-bold info-item-value d-block pull-left pr-25')[0].text                  
                batnum.append(num_baths)
            except:
                break        
        
        # For loop to grab listing status
        for k in range(0,40):
            try:
                listing_status = soup.find_all(class_='listing-statusd-block pull-left pl-50 pr-1 status-marker status-closed')[k].text
                status.append(listing_status)
            except:
                break        
        
        # For loop to grab the square footage
        for j in range(0,80,2):
            try:
                sq_footage = soup.find_all(class_='listing-info-item font-size-sm line-height-base d-block pull-left pr-50 listing-sqft last')[j].find_all(class_='font-weight-bold info-item-value d-block pull-left pr-25')[0].text
                sqfoot.append(sq_footage)
            except:
                break
        
        # For loop to grab the built year
        for j in range(1,81,2):
            try:
                year_built = soup.find_all(class_='listing-info-item font-size-sm line-height-base d-block pull-left pr-50 listing-sqft last')[j].find_all(class_='font-weight-bold info-item-value d-block pull-left pr-25')[0].text
                buildt.append(year_built)
            except:
                break
        
        # For loop to grab the name of broker agency
        for k in range(0,40):
            try:
                listing_broker = soup.find_all(class_='listing-info clearfix font-size-sm line-height-base mb-25 listing-broker')[k].find_all(class_='font-weight-bold d-block pull-left')[0].text
                broker.append(listing_broker)
            except:
                break
```

```python
# Run the scrape function for MLS San Francisco (Sold)
import time
start = time.time()
mls_sfsold_data(30)
end = time.time()
```

```python
# # Check the length of arrays (scaped values)
# print(len(titles),len(prices),len(mlsnum),len(bednum),len(batnum),len(status),len(sqfoot),len(buildt),len(broker))

# Check scrape run time
print(start-end,"seconds")
mls_sfsold_df = pd.DataFrame({'Address': titles, 'MLS_Number': mlsnum, 'Prices': prices, 'Bedrooms': bednum, 'Bathrooms': batnum, 'Size': sqfoot, 'Status': status, 'Built_yr': buildt, 'Broker': broker})
```
